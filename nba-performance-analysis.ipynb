```python
import pandas as pd
import matplotlib.pyplot as plt
import os


# Load the dataset from URL or file
url = 'https://raw.githubusercontent.com/annlntv/sg_da/main/NBA_2024_cleaned.csv'
df = pd.read_csv(url)

# Define high scorers
high_scorers = df[df['PTS'] >= 20]

# Metrics for comparison
metrics = ['FGA', '3PA', 'FT', 'AST', 'TRB']

# Calculate mean values for high scorers and others
mean_values = {
    'High Scorers': high_scorers[metrics].mean(),
    'Other Players': df[df['PTS'] < 20][metrics].mean()
}

# Create a DataFrame for easy comparison
mean_comparison = pd.DataFrame(mean_values)

# Plot the comparison
plt.figure(figsize=(10, 6))
mean_comparison.plot(kind='bar', width=0.8)
plt.title('Comparison of Metrics: High Scorers vs Other Players')
plt.ylabel('Mean Value')
plt.xlabel('Metrics')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot as an image
plt.savefig('high_scorers_vs_others.png')
print(os.getcwd())  # This will print the current working directory

plt.show()
```
